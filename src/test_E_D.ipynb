```python
# viết hàm chuyển một chuỗi có dạng '[6, 1, 3, 2]' về thành mảng trong numpy
import numpy as np

def str_to_arr(str):
    return np.array(eval(str))
```

```python
def top_n_indices(arr, n):
    # Chuyển danh sách thành mảng NumPy
    arr_np = np.array(arr)
    
    # Sử dụng np.argsort() để sắp xếp mảng theo thứ tự giảm dần của các phần tử
    sorted_indices = np.argsort(arr_np)[::-1]
    
    # Lấy chỉ mục của n phần tử lớn nhất
    top_n_indices = sorted_indices[:n]
    
    return top_n_indices

# Kiểm tra hàm với một danh sách và n
arr = [1, 3, 5, 2, 4]
n = 3
result_indices = top_n_indices(arr, n)
print("Chỉ mục của", n, "phần tử lớn nhất trong danh sách là:", result_indices)
```
